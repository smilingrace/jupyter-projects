```python
import numpy as np
import matplotlib.pyplot as plt

# Time vector
t = np.linspace(0, 2*np.pi, 1000)  # One period, assuming T=1s

# Amplitudes and phase difference
A1 = 1  # Amplitude of oscillation 1
A2 = 2 * A1  # Amplitude of oscillation 2
phase_difference = np.pi / 4  # Phase difference between oscillation 2 and oscillation 1

# Oscillation equations
x1 = A1 * np.cos(t)
x2 = A2 * np.cos(t - phase_difference)  # Phase-shifted oscillation 2

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t, x1, label='Oscillation 1 (A1)')
plt.plot(t, x2, label='Oscillation 2 (2*A1)')

plt.title('Comparison of Two Oscillations')
plt.xlabel('Time (s)')
plt.ylabel('Displacement')

plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Time vector
t = np.linspace(0, 1, 1000)  # One period, assuming T=1s

# Amplitudes and phase difference
A1 = 1  # Amplitude of oscillation 1
A2 = 2 * A1  # Amplitude of oscillation 2
phase_difference = np.pi / 4  # Phase difference between oscillation 2 and oscillation 1

# Oscillation equations
x1 = A1 * np.cos(2*np.pi*t)
x2 = A2 * np.cos(2*np.pi*t - phase_difference)  # Phase-shifted oscillation 2

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t, x1, label='Oscillation 1')
plt.plot(t, x2, label='Oscillation 2')

plt.title('Comparison of Two Oscillations')
plt.xlabel('Time (s)')
plt.ylabel('Displacement')

plt.legend()
plt.grid(True)
plt.show()

# Time vector
t = np.linspace(0, 2*np.pi, 1000)  # One period, assuming omega_0 = 1

# Parameters
Q = 1  # Amplitude of charge
omega_0 = 1  # Angular frequency

# Charge function
q = Q * np.cos(omega_0 * t)

# Current function (derivative of charge function)
I = -Q * omega_0 * np.sin(omega_0 * t)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t, q, label='Charge (q)')
plt.plot(t, I, label='Current (I)')

plt.title('Charge and Current vs Time')
plt.xlabel('Time (s)')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
