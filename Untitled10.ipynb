```python
import numpy as np
import matplotlib.pyplot as plt

hbar = 1  
m = 1     
k_values = np.linspace(-5, 5, 100)
E_values = (hbar**2 * k_values**2) / (2 * m)

plt.plot(k_values, E_values)
plt.title('Dispersion Relation for a 1D Free Particle')
plt.xlabel('Wave Number (k)')
plt.ylabel('Energy (E)')
plt.grid(True)
plt.show()
```
