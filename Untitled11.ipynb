```python
import numpy as np
import matplotlib.pyplot as plt

def rectangular_approximation(x, a, epsilon):
    return (1/(2*epsilon)) * (np.heaviside(x - a + epsilon/2, 1) - np.heaviside(x - a - epsilon/2, 1))

a = 0
epsilon_values = [0.2, 0.1, 0.05, 0.01]
x = np.linspace(-1, 1, 200)

plt.figure(figsize=(8, 6))
for epsilon in epsilon_values:
    plt.plot(x, rectangular_approximation(x, a, epsilon), label=f'$\epsilon={epsilon}$')

plt.title('Rectangular Approximation of $\delta(x - a)$')
plt.xlabel('$x$')
plt.ylabel('$\delta_{\epsilon}(x - a)$')
plt.axvline(x=a, color='r', linestyle='--', label='$a$')
plt.legend()
plt.grid(True)
plt.show()
```
