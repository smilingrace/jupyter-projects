```python
import matplotlib.pyplot as plt
import numpy as np

I=np.array([95.4, 88.4, 84.9, 82, 71.7, 70.9, 65.3])
V_AB=np.array([0.652, 0.653, 0.654, 0.655, 0.656, 0.657, 0.658])
xp=np.arange(0.651, 0.659, 0.001)
best_fit_param=np.polyfit(V_AB,I,1)
print(best_fit_param)
y_best_fit=best_fit_param[1]+best_fit_param[0]*xp
plt.title('Current Voltage Plot for Thevenin Resistance')
plt.ylabel('Current (microAmps)')
plt.xlabel('Voltage (Volts)')
caption="Estimated error: +/- 0.05 microA   +/- 0.0005V"

plt.plot(xp,y_best_fit, label="y=-0.00494643x-0.00331971")
plt.figtext(0,0, caption, wrap=True, horizontalalignment='center', fontsize=12)
plt.legend()
plt.scatter(V_AB,I)
plt.plot(xp,y_best_fit, label="y=")
plt.show()
```
