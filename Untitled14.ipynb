```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
m = 1.0  # mass
L = 1.0  # length of pendulum
g = 9.8  # acceleration due to gravity
a = 1.0  # spacing between pendulums
k_values = np.linspace(-np.pi / a, np.pi / a, 1000)  # wave number values

# Dispersion relation
def dispersion_relation(k, m, L, g, a, kappa):
    return np.real(np.sqrt(-(g + 4 * kappa * np.sin(k * a / 2)**2) / (m * L)))

# Plotting
kappa_values = [1.0, 5.0, 10.0]  # try different values of kappa

plt.figure(figsize=(8, 6))

for kappa in kappa_values:
    omega_values = dispersion_relation(k_values, m, L, g, a, kappa)
    plt.plot(k_values, omega_values, label=f'κ = {kappa}')

plt.axhline(y=0, color='black', linestyle='--', linewidth=0.8)
plt.axvline(x=0, color='black', linestyle='--', linewidth=0.8)
plt.title('Dispersion Relation for Coupled Pendulums')
plt.xlabel('Wave Number (k)')
plt.ylabel('Angular Frequency (ω)')
plt.legend()
plt.grid(True)
plt.show()
```
