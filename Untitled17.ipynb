```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
hbar = 1.0  # Planck's constant over 2*pi
m = 1.0     # Mass of the particle
omega = 1.0 # Oscillator frequency
F = 0.1     # Weak constant force

# Unperturbed ground state wave function for harmonic oscillator
def psi_0(x):
    return (m * omega / (np.pi * hbar))**(1/4) * np.exp(-m * omega * x**2 / (2 * hbar))

# Exact ground state wave function from previous homework (replace with your function)
def psi_g(x):
    return (m * omega / (np.pi * hbar))**(1/4) * np.exp(-3*m * omega * x**2 / (2 * hbar))

# Range of x values
x_values = np.linspace(-5, 5, 1000)

# Calculate wave functions
psi_0_values = psi_0(x_values)
psi_g_values = psi_g(x_values)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x_values, psi_0_values, label=r'$\psi_0(x)$ (Perturbation Theory)')
plt.plot(x_values, psi_g_values, label=r'$\psi_g(x)$ (Exact)')
plt.xlabel('x')
plt.ylabel(r'$\psi(x)$')
plt.title('Ground State Wave Function')
plt.legend()
plt.grid(True)
plt.show()
```
