```python
import numpy as np
from matplotlib import pyplot as plt
import math

#LED Current I (mA)
I_l=np.array([0,0,5.56,8.03,8.74,9.13,9.4,9.52,9.46,8.62,7.9])
#Base Current I (mA)
I_b=np.array([18.95,18.46,4.99,2.32,1.52,1.1,0.77,0.57,0.44,0.39,0.35])
#Base Voltage V
V_b=np.array([0.807,0.808,0.767,0.74,0.726,0.715,0.703,0.694,0.686,0.68,0.676])

plt.ylim(-1,11)
plt.title("Plot for Transistor Gain (I_l vs I_b)")
plt.xscale("log")
plt.ylabel("LED Current mA (I)")
plt.xlabel("Base Current mA (I)")
plt.scatter(I_b,I_l)
```

```python
plt.ylim(-1,11)
plt.title("LED Current vs Base-Emitter Voltage (I_l vs V_b)")
plt.xscale("log")
plt.ylabel("LED Current mA (I)")
plt.xlabel("Base Voltage (V)")
plt.scatter(V_b,I_l)
```
