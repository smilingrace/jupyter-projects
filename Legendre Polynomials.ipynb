```python
import sympy as sp

z = sp.Symbol('z')
f = sp.sin(sp.pi * z)

# Compute the Legendre polynomial expansion up to degree n
n = 10  # Change this value to adjust the number of terms
legendre_expansion = sp.series(f, z, n=n)

print("Legendre polynomial expansion:")
print(legendre_expansion)
```

```python
import sympy as sp
from sympy import sin
from sympy.series.sequences import RecursiveSeq
from sympy.series.order import Order

z = sp.Symbol('z')
f = sin(sp.pi * z)

n = 9  
legendre_expansion = sp.series(f, z, n=n).removeO()

print("Legendre polynomial expansion:")
print(legendre_expansion)

import numpy as np
import matplotlib.pyplot as plt

z_vals = np.linspace(-1, 1, num=1000)

f_approx = sp.lambdify(z, legendre_expansion, "numpy")

f_approx_vals = f_approx(z_vals)

f_exact_vals = np.sin(np.pi * z_vals)

plt.plot(z_vals, f_exact_vals, label='Exact function')
plt.plot(z_vals, f_approx_vals, label='Approximation')
plt.xlabel('z')
plt.ylabel('f(z)')
plt.legend()
plt.show()
```

```python
import sympy as sp
from sympy import sin
from sympy.series.sequences import RecursiveSeq
from sympy.series.order import Order

z = sp.Symbol('z')
g = sin(3 * sp.pi * z)

n = 20  
legendre_expansion = sp.series(g, z, n=n).removeO()

print("Legendre polynomial expansion:")
print(legendre_expansion)

import numpy as np
import matplotlib.pyplot as plt

z_vals = np.linspace(-1, 1, num=100)

g_approx = sp.lambdify(z, legendre_expansion, "numpy")

g_approx_vals = g_approx(z_vals)

g_exact_vals = np.sin(3 * np.pi * z_vals)

plt.plot(z_vals, g_exact_vals, label='Exact function')
plt.plot(z_vals, g_approx_vals, label='Approximation')
plt.xlabel('z')
plt.ylabel('g(z)')
plt.legend()
plt.show()
```
