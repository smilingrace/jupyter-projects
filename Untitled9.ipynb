```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate S/N in the source-limited case
def sn_source_limited(N_star, N_sky):
    return np.sqrt(N_star) / np.sqrt(N_star + N_sky)

# Function to calculate S/N in the background-limited case
def sn_background_limited(N_star, N_sky):
    return N_star / np.sqrt(N_sky)

# Generate values for exposure time (t)
t_values = np.linspace(0.1, 10, 100)

# Set values for counts from the star and counts from the sky
N_star = 1000
N_sky = 100

# Calculate S/N for each case
sn_source = [sn_source_limited(N_star, N_sky) for _ in t_values]
sn_background = [sn_background_limited(N_star, N_sky) for _ in t_values]

# Plot the results
plt.figure(figsize=(8, 6))
plt.plot(t_values, sn_source, label='Source-limited')
plt.plot(t_values, sn_background, label='Background-limited')
plt.xlabel('Exposure Time (t)')
plt.ylabel('Signal-to-Noise Ratio (S/N)')
plt.legend()
plt.title('S/N vs. Exposure Time')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate S/N in the source-limited case
def sn_source_limited(N_star, N_sky, t):
    return np.sqrt(N_star * t) / np.sqrt(N_star * t + N_sky)

# Function to calculate S/N in the background-limited case
def sn_background_limited(N_star, N_sky, t):
    return N_star * t / np.sqrt(N_sky)

# Generate values for exposure time (t)
t_values = np.linspace(0.1, 10, 100)

# Set values for counts from the star and counts from the sky
N_star = 1000
N_sky = 100

# Calculate S/N for each case with varying exposure time
sn_source = [sn_source_limited(N_star, N_sky, t) for t in t_values]
sn_background = [sn_background_limited(N_star, N_sky, t) for t in t_values]

# Plot the results
plt.figure(figsize=(8, 6))
plt.plot(t_values, sn_source, label='Source-limited')
plt.plot(t_values, sn_background, label='Background-limited')
plt.xlabel('Exposure Time (t)')
plt.ylabel('Signal-to-Noise Ratio (S/N)')
plt.legend()
plt.title('S/N vs. Exposure Time')
plt.grid(True)
plt.show()
```
