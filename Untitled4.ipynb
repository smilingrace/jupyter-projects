```python
import numpy as np
import matplotlib.pyplot as plt

r=5
x=np.linspace(0,2*np.pi, 100)
y=(np.sqrt(27))/(4*np.pi*r*(2+np.cos(3*x)))

plt.plot(x,y, color="blue", label="Wave Function")
plt.plot(x,y**2, color="red", label="Prob Dens")
plt.xlabel("Angle Phi")
plt.ylabel("Amplitude")
plt.legend()
```
