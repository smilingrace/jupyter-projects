```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
E0 = 1.0
R = 1.0
V0 = 2.0  # You can choose any value for V0

# Define the potential function
def potential(theta):
    return -E0 * (R - R**3 / R**2) * np.cos(theta) + E0 * R + V0

# Generate theta values
theta_values = np.linspace(0, 2 * np.pi, 1000)

# Calculate potential for each theta
potential_values = potential(theta_values)

# Create a 2D array for contour plotting
theta_grid, potential_grid = np.meshgrid(theta_values, potential_values)

# Plotting
plt.figure(figsize=(8, 6))
contour = plt.contourf(theta_grid, potential_grid, levels=100, cmap='viridis')
plt.title('2D Contour Plot of Electric Potential')
plt.xlabel('Theta (radians)')
plt.ylabel('Electric Potential')
plt.colorbar(contour, label='Potential')
plt.show()
```
