```python
import matplotlib.pyplot as plt
import numpy as np

TN = 6.1
TC = 77
Ni0 = 1000
Co0 = 0

time = np.arange(0, 200, 0.1)
Ni = np.zeros_like(time)
Co = np.zeros_like(time)

Ni[0] = Ni0
Co[0] = Co0

for i in range(1, len(time)):
    dNi_dt = -0.113 * Ni[i-1]
    dCo_dt = 0.113 * Ni[i-1] - (Ni0/Ni0 + Co0) * Co[i-1] / TC * np.log(2)
    Ni[i] = Ni[i-1] + 0.1 * dNi_dt
    Co[i] = Co[i-1] + 0.1 * dCo_dt
    
plt.plot(time, Co)
plt.xlabel('Time (days)')
plt.ylabel('Number of atoms')
plt.show()
```
