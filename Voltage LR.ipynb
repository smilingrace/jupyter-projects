```python
import matplotlib.pyplot as plt
import numpy as np
V_AB = np.arange(0., 5., 0.2)
V_S = 3.3
R_TH = 100
I = (V_S/(5*R_TH))-(V_AB/R_TH)

plt.ylabel('Current Through Load Resistor (Amps)')
plt.xlabel('Voltage Across Load Resistor (Volts)')
#plt.ylim([0,.0066])

plt.plot (V_AB, I)
plt.show()
```
