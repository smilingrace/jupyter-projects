```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants and parameters
V0 = 1.0  # Adjust this as needed
a = 1.0
b = 1.0

# Function to calculate the potential at a specific point (x, y)
def potential(x, y, n_max=10):
    result = 0
    for n in range(1, n_max + 1):
        k = (2 * n - 1) * np.pi / (2 * b)
        result += np.cosh(k * y) / np.cosh((np.pi * b) / 2) * np.sin((np.pi * x * (2 * n - 1)) / (2 * a))
    return V0 * result

# Create a meshgrid for x and y
x_vals = np.linspace(0, a, 100)
y_vals = np.linspace(0, b, 100)
x, y = np.meshgrid(x_vals, y_vals)

# Calculate the potential for each point in the meshgrid
z = potential(x, y)

# Plot the potential surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Label axes
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('Potential')

# Show the plot
plt.show()
```
