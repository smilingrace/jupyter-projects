```python
import numpy as np
import matplotlib.pyplot as plt

# Define the angle for the filled phasor
angle_filled = np.pi / 4  # Adjust this angle as needed

# Define the angles for the phasors for each scenario
angles_scenario_1 = [0, np.pi / 2, np.pi]  # Angles for V0, VC, and VL
angles_scenario_2 = [0, angle_filled, angle_filled + np.pi]  # Angles for V0, VC, and VL
angles_scenario_3 = [0, np.pi / 2, np.pi]  # Angles for V0, VC, and VL

# Define the magnitudes for the phasors for each scenario
magnitude = 1  # Adjust this magnitude as needed
magnitudes_scenario_1 = [magnitude, magnitude * np.cos(angle_filled), magnitude * np.sin(angle_filled)]
magnitudes_scenario_2 = [magnitude, magnitude, magnitude]
magnitudes_scenario_3 = [magnitude, magnitude * np.cos(angle_filled), magnitude * np.sin(angle_filled)]

# Plot phasor diagrams for each scenario
plt.figure(figsize=(12, 4))

# Scenario 1
plt.subplot(1, 3, 1, projection='polar')
plt.title('Scenario 1: $  _0$')
plt.plot(angles_scenario_1, magnitudes_scenario_1, 'bo-', label='$V_0$')
plt.plot([angles_scenario_1[1], angles_scenario_1[2]], [magnitudes_scenario_1[1], magnitudes_scenario_1[2]], 'ro-', label='$V_C$')
plt.plot([angles_scenario_1[0], angles_scenario_1[2]], [magnitudes_scenario_1[0], magnitudes_scenario_1[2]], 'go-', label='$V_L$')
plt.legend(loc='upper right')

# Scenario 2
plt.subplot(1, 3, 2, projection='polar')
plt.title('Scenario 2: $ = _0$')
plt.plot(angles_scenario_2, magnitudes_scenario_2, 'bo-', label='$V_0$')
plt.plot([angles_scenario_2[1], angles_scenario_2[2]], [magnitudes_scenario_2[1], magnitudes_scenario_2[2]], 'ro-', label='$V_C$')
plt.plot([angles_scenario_2[0], angles_scenario_2[2]], [magnitudes_scenario_2[0], magnitudes_scenario_2[2]], 'go-', label='$V_L$')
plt.legend(loc='upper right')

# Scenario 3
plt.subplot(1, 3, 3, projection='polar')
plt.title('Scenario 3: $  _0$')
plt.plot(angles_scenario_3, magnitudes_scenario_3, 'bo-', label='$V_0$')
plt.plot([angles_scenario_3[1], angles_scenario_3[2]], [magnitudes_scenario_3[1], magnitudes_scenario_3[2]], 'ro-', label='$V_C$')
plt.plot([angles_scenario_3[0], angles_scenario_3[2]], [magnitudes_scenario_3[0], magnitudes_scenario_3[2]], 'go-', label='$V_L$')
plt.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the resonant frequency and quality factor
resonant_frequency = 100e6  # 100 MHz
quality_factor = 500

# Calculate the spacing of broadcast frequencies
spacing = resonant_frequency / quality_factor

# Define the frequencies for two adjacent stations
station_a = resonant_frequency - spacing / 2
station_b = resonant_frequency + spacing / 2

# Plotting the graph
plt.figure(figsize=(10, 4))
plt.eventplot([station_a, station_b], orientation='horizontal', linelengths=0.8, colors='b')
plt.text(station_a - 0.2e6, 0.1, 'Station A', horizontalalignment='right', verticalalignment='center')
plt.text(station_b + 0.2e6, 0.1, 'Station B', horizontalalignment='left', verticalalignment='center')
plt.xlabel('Frequency (MHz)')
plt.title('Spacing of Broadcast Frequencies for FM Stations')
plt.grid(axis='y')
plt.ylim(0, 1)
plt.xlim(station_a - 0.5e6, station_b + 0.5e6)
plt.tight_layout()
plt.show()
```
