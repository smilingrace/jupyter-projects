```python
import numpy as np
import matplotlib.pyplot as plt

# Circuit parameters
L = 1  # Inductance (H)
R = 50  # Resistance (Î©)
C = 2e-3  # Capacitance (F)
V0 = 10  # Initial voltage across the capacitor (V)
tau = L / R  # Damping time (s)
# Quality factor (Q-factor)
Q = np.sqrt(L / (C * R))

# Damped angular frequency (approximation)
omega_d = omega0 * (1 - 1 / (2 * Q**2))
omega0 = 1 / np.sqrt(L * C)  # Natural angular frequency (rad/s)
omega_d = np.sqrt(omega0**2 - (1 / tau**2))  # Damped angular frequency (rad/s)
omega_d_squared = omega0**2 - (1 / tau**2)
if omega_d_squared < 0:
    omega_d = 0
else:
    omega_d = np.sqrt(omega_d_squared)
# Time vector
t = np.linspace(0, 10, 1000)  # 10 cycles

# Charge on the capacitor
Q = V0 * C

# Current through the circuit
I0 = V0 / R  # Initial current
current = I0 * np.exp(-t / tau) * np.sin(omega_d * t)

# Current derivative
current_derivative = (-I0 / tau) * np.exp(-t / tau) * np.sin(omega_d * t) + I0 * np.exp(-t / tau) * omega_d * np.cos(omega_d * t)

# Plotting
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(t, Q * np.exp(-t / tau), label='Charge (Q)')
plt.xlabel('Time (s)')
plt.ylabel('Charge (C)')
plt.title('Charge on the Capacitor')

plt.subplot(3, 1, 2)
plt.plot(t, current, label='Current (I)')
plt.xlabel('Time (s)')
plt.ylabel('Current (A)')
plt.title('Current in the Circuit')

plt.subplot(3, 1, 3)
plt.plot(t, current_derivative, label='Current Derivative (dI/dt)')
plt.xlabel('Time (s)')
plt.ylabel('Current Derivative (A/s)')
plt.title('Current Derivative')

plt.tight_layout()
plt.show()
```
