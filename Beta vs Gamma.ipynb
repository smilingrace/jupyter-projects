```python
import matplotlib
import numpy as np

from matplotlib import pyplot as plt

c=3*10**8 #speed of light in m/s

G=np.arange(1,1000,0.1) #Velocity parameter in meters per second
B=(1-(1/(G**2)))**0.5 #

plt.plot(G,B)
plt.xlim([1,10])
plt.ylim([0,1])
plt.title('Beta as a Function of Gamma')
plt.ylabel('Beta')
plt.xlabel('Gamma')
plt.show()
```
